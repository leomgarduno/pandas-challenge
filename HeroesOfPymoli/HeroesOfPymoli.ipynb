```python
#import dependencies
import pandas as pd
import numpy as np 

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

```python
#display the total number of players

totalplayers = len(purchase_df["SN"].unique())

pd.DataFrame({"Total Players" : [totalplayers]})
```

```python
#Purchasing Analysis (Total)
 
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results

uniqueitems = len(purchase_df["Item ID"].unique())
averageprice = purchase_df["Price"].mean()
countpurchases = len(purchase_df["Purchase ID"].unique())
totalrevenue = purchase_df["Price"].sum()

#Generate DataFrame  & Format Data

purchase_data_analysis_df = pd.DataFrame({"Number of Unique Items" : [uniqueitems], 
                                            "Average Price" : ['${:,.2f}'.format(averageprice)], 
                                            "Number of Purchases" : countpurchases,
                                           "Total Revenue" : ['${:,.2f}'.format(totalrevenue)]})

#Display df
purchase_data_analysis_df
```

```python
##Gender Demographics
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

#Calculate groupby gender
groupgender = purchase_df.groupby("Gender")

#Calculate nunique players
gendercount = groupgender.nunique()["SN"]
gendercount

#calculate percentages
percentage_of_gender = ((gendercount / totalplayers)*100)
percentage_of_gender

#send to a df
gender_df = pd.DataFrame({"Total Count" : gendercount, 
                          "Percentage of Players" : (percentage_of_gender)})

#sort table
gender_df = gender_df.sort_values(["Percentage of Players"], ascending = False)

#format df
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}%".format)


#Display df
gender_df
```

```python
#Purchasing Analysis (Gender)
 
#Run basic calculations to obtain 

#purchase count
purchase_count = groupgender["Purchase ID"].count()
purchase_count

#avg. purchase price 
avg_purchase_price = groupgender["Price"].mean()
avg_purchase_price

#avg. purchase total per person 
total_purchase = groupgender["Price"].sum()
total_purchase

#avg total purchase per person
avg_total_purchase = total_purchase/gendercount
avg_total_purchase

#Create a summary data frame to hold the results

purchase_analysis_gender_df = pd.DataFrame({"Purchase Count" : purchase_count, 
                          "Avg Purchase Price" : avg_purchase_price,
                                   "Total Purchase Value" : total_purchase,
                                   "Avg Total Purchase Per Person" : (avg_total_purchase)})

#Optional: give the displayed data cleaner formatting
purchase_analysis_gender_df["Avg Purchase Price"] = purchase_analysis_gender_df["Avg Purchase Price"].map("${:.2f}".format)
purchase_analysis_gender_df["Total Purchase Value"]  = purchase_analysis_gender_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_gender_df["Avg Total Purchase Per Person"] = purchase_analysis_gender_df["Avg Total Purchase Per Person"].map("${:.2f}".format)

#Display the summary data frame
purchase_analysis_gender_df
```

```python
#Age Demographics
#Establish bins for ages

maxage = purchase_df["Age"].max()
bins = [0, 9, 14, 19, 24, 29, 34, 39, maxage]
group_names = ["<10", "10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Categorize the existing players using the age bins. Hint: use pd.cut()

purchase_df["Age Range"] = pd.cut(purchase_df["Age"], bins, labels=group_names)
purchase_df.head()

#Calculate groupby age
groupage = purchase_df.groupby("Age Range")
groupage

#Calculate nunique players
ages_count = groupage.nunique()["SN"]
ages_count

#Calculate the numbers and  percentages by age group
age_percentage = ages_count/totalplayers *100
age_percentage

#Create a summary data frame to hold the results

age_ranges_df = pd.DataFrame({"Total Count" : ages_count, 
                          "Percentage Of Players" : age_percentage})

#Optional: round the percentage column to two decimal points
age_ranges_df["Percentage Of Players"] = age_ranges_df["Percentage Of Players"].map("{:.2f}%".format)

#Display Age Demographics Table
age_ranges_df
```

```python
#Purchasing Analysis (Age)
#Bin the purchase_data data frame by age

purchase_df["Age Range"] = pd.cut(purchase_df["Age"], bins, labels=group_names)
purchase_df.head()

#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below

#purchase count
purchase_count_age = groupage["Purchase ID"].count()
purchase_count_age

#avg. purchase price 
avg_purchase_price_age = groupage["Price"].mean()
avg_purchase_price_age

#avg. purchase total per person 
total_purchase_age = groupage["Price"].sum()
total_purchase_age

#avg total purchase per person
avg_total_purchase_age = total_purchase_age/ages_count
avg_total_purchase_age


#Create a summary data frame to hold the results

purchase_analysis_ages_df = pd.DataFrame({"Purchase Count" : purchase_count_age, 
                          "Avg Purchase Price" : avg_purchase_price_age,
                                   "Total Purchase Value" : total_purchase_age,
                                   "Avg Total Purchase Per Person" : avg_total_purchase_age})

#Optional: give the displayed data cleaner formatting  #REVIEW

purchase_analysis_ages_df["Avg Purchase Price"] = purchase_analysis_ages_df["Avg Purchase Price"].map("${:.3}".format)
purchase_analysis_ages_df["Total Purchase Value"]  = purchase_analysis_ages_df["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_ages_df["Avg Total Purchase Per Person"] = purchase_analysis_ages_df["Avg Total Purchase Per Person"].map("${:.2f}".format)

#Display the summary data frame
purchase_analysis_ages_df
```

```python
#Top Spenders
#Run basic calculations to obtain the results in the table below

#Calculate groupby spender
groupspender = purchase_df.groupby("SN")
groupspender

#Purchase Count by spender
purchase_count_spender = groupspender["Purchase ID"].count()


#Avg Purchas Price by Spender
avg_purchase_spender = groupspender["Price"].mean()


#Total Purchase Value by Spender
total_purchase_spender = groupspender["Price"].sum()

    
#Create a summary data frame to hold the results
spender_df = pd.DataFrame({"Purchase Count" : purchase_count_spender, 
                           "Avg Purchase Price" : avg_purchase_spender,
                           "Total Purchase Value" : total_purchase_spender})

#Sort the total purchase value column in descending order

spender_sorted_df = spender_df.sort_values(["Purchase Count"], ascending = False)

#Optional: give the displayed data cleaner formatting #REVIEW

spender_sorted_df["Avg Purchase Price"] = spender_sorted_df["Avg Purchase Price"] .map("${:.3}".format)
spender_sorted_df["Total Purchase Value"] = spender_sorted_df["Total Purchase Value"] .map("${:.4}".format)

#Display a preview of the summary data frame
spender_sorted_df.head()
```

```python
#Most Popular Items
#Retrieve the Item ID, Item Name, and Item Price columns

popularitems_df = purchase_df[["Item ID", "Item Name", "Price"]]
popularitems_df

#Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value
itemsgroup = popularitems_df.groupby(["Item ID", "Item Name"])
itemsgroup

#Purchase Count by Item
purchase_count_items = itemsgroup["Item ID"].count()

#Avg Purchas Price by Item
avg_purchase_items = itemsgroup["Price"].mean()

#Total Purchase Value by Item
total_purchase_items = itemsgroup["Price"].sum()
    
#Create a summary data frame to hold the results
popular_items_df = pd.DataFrame({"Purchase Count" : purchase_count_items, 
                           "Avg Purchase Price" : avg_purchase_items,
                           "Total Purchase Value" : total_purchase_items})



#Sort the purchase count column in descending order
popular_items_sorted_purchase_df = popular_items_df.sort_values(["Purchase Count"], ascending = False)

#Optional: give the displayed data cleaner formatting

popular_items_sorted_purchase_df["Avg Purchase Price"]  = popular_items_sorted_purchase_df["Avg Purchase Price"].map("${:,.2f}".format)
popular_items_sorted_purchase_df["Total Purchase Value"] = popular_items_sorted_purchase_df["Total Purchase Value"].map("${:.2f}".format)


#Display a preview of the summary data frame
popular_items_sorted_purchase_df.head()
```

```python
#Most Profitable Items
#Sort the above table by total purchase value in descending order
popular_items_sorted_purchase_df = popular_items_df.sort_values(["Total Purchase Value"], ascending = False)

#Optional: give the displayed data cleaner formatting

popular_items_sorted_purchase_df["Avg Purchase Price"]  = popular_items_sorted_purchase_df["Avg Purchase Price"].map("${:,.2f}".format)
popular_items_sorted_purchase_df["Total Purchase Value"] = popular_items_sorted_purchase_df["Total Purchase Value"].map("${:.2f}".format)



#Display a preview of the data frame
popular_items_sorted_purchase_df.head()
```
